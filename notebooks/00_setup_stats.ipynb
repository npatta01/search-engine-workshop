```python
import pandas as pd
```

```python
path = "../data/raw/conversions.tsv000"
```

```python
df = pd.read_csv(path,sep="\t")
```

```python
df.head()
```

```python
df_res = (df.groupby(["keyword"], as_index=False).size().sort_values("size", ascending=False))
```

```python
df_res["num_keywords"] = df_res["keyword"].apply(lambda x: len(x.split(" ")))
```

```python
df_res = df_res[(df_res["num_keywords"] > 1) & (df_res["size"] > 1)]
```
